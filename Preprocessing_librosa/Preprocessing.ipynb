```python
import librosa,librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
file="The north wind and the sun.wav"
```

```python
#waveform
signal,sampling_rate=librosa.load(file)

librosa.display.waveplot(signal,sampling_rate)

#time domain
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
#fft waveform -> spectrum

fft=np.fft.fft(signal)
magnitude=np.abs(fft)
frequency=np.linspace(0,sampling_rate,len(magnitude))

ranged_magnitude=magnitude[:int(len(magnitude)/5)]
ranged_frequency=frequency[:int(len(frequency)/5)]

plt.plot(ranged_frequency,ranged_magnitude)
#frequency domain
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
#stft spectrum->spectrogram

#number of samples
window_size=int(sampling_rate/50) #20ms
#frame shifting
frame_shift=int(sampling_rate/100) #10ms

stft=librosa.core.stft(signal,hop_length=frame_shift,n_fft=window_size)
spectrogram=np.abs(stft)

log_spectrogram=librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram,sr=sampling_rate,hop_length=frame_shift)

plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.show()
```

```python
#MFCC

MFCC=librosa.feature.mfcc(signal, n_fft=window_size,hop_length=frame_shift, n_mfcc=39)

librosa.display.specshow(MFCC,sr=sampling_rate,hop_length=frame_shift)

plt.xlabel("Time")
plt.ylabel("MFCC")
plt.colorbar()
plt.show()
```
