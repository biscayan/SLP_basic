```python
import struct

wav_file= open ('PBW1002.WAV','rb')
print(wav_file)
print(type(wav_file))
```

```python

wav_file.seek(0,2)
file_size=wav_file.tell()
print(file_size,"bytes")

wav_file.seek(44,0)
file_head=wav_file.tell()
print(file_head)

sample_size=file_size-file_head
print(sample_size)
```

```python
data_list=[]
byte=wav_file.read(2)

while byte:
    byte=wav_file.read(2)
    data_list.append(byte)
    
print(data_list)    
print(len(data_list))
```

```python
out_file=open('PBW1002_sampling_data.txt','w')
for i in data_list:
    unpacked=struct.unpack('<h',i)
    print(unpacked[0])
    out_file.write("{0}\n".format(unpacked[0]))
out_file.close()
```
